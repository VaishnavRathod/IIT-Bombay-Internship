```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('C:/Users/VAISHNAV/Downloads/AH_Monthly_Provisional_Counts_of_Deaths_by_Age_Group_and_HHS_region_for_Select_Causes_of_Death__2019-2021.csv')
data
```

```python
data.info()
```

## Natural Cause Deaths in US: 15-24y

```python
data_15_24 = data[(data['AgeGroup'] == '15-24 years') & 
     (data['HHSRegion'] == 'United States')][['NaturalCause','Date Of Death Year','Date Of Death Month']].reset_index(drop=True)
#data_15_24
```

```python
plt.figure(figsize=(7,4))
sns.set_palette(palette='bright')
sns.lineplot(x=data_15_24['Date Of Death Month'],y=data_15_24['NaturalCause'], 
             hue=data_15_24['Date Of Death Year'], palette=['blue', 'orange', 'red'])
plt.legend(loc='upper center', bbox_to_anchor=(0.5, 1.15), ncol = 3)
plt.grid(True, alpha = 0.5)
plt.tight_layout()
plt.xlim([1,12])
plt.xticks(np.arange(1,13))
plt.xlabel('Month')
plt.ylabel('Natural cause deaths(15-24y)')
plt.ylim([0,1500])
plt.yticks([0,500,1000,1500])
plt.show()
```
