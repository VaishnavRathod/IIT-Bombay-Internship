```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.dates as m_dates
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('owid-covid-data.csv')
data.head()
```

```python
data.info()
```

```python
data.shape
```

```python
data['iso_code'].unique()
```

```python
data_ind = data[data['iso_code'] == 'IND'] 
```

```python
data_ind.head()
```

```python
data_ind['date'] = pd.to_datetime(data_ind['date'])
```

```python
data_ind.info()
```

```python
data_ind_time = data_ind[(data_ind['date'] >= '2020-03-01') & (data_ind['date'] <= '2021-05-10')].reset_index(drop=True)
```

```python
cases = data_ind_time[['date','new_cases_per_million']]
deaths = data_ind_time[['date','new_deaths_per_million']]
```

```python
cases['total_cases_per_million(7_day_avg)'] = cases['new_cases_per_million'].rolling(7).mean()
deaths['total_deaths_per_million(7_day_avg)'] = deaths['new_deaths_per_million'].rolling(7).mean()
```

### Plotting graph

```python
plt.figure(figsize=(10,5))
plt.plot(cases['date'], cases['total_cases_per_million(7_day_avg)']/100,
         label='Daily cases (7-day avg) per million [X 100]')
plt.plot(deaths['date'], deaths['total_deaths_per_million(7_day_avg)'],
         label='Daily deaths (7-day avg) per million', color='red')
plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.05), ncol = 2)
myfmt=m_dates.DateFormatter("1-%b-20%y")
plt.gca().xaxis.set_major_formatter(myfmt)
plt.gca().xaxis.set_major_locator(m_dates.MonthLocator(interval=3))
plt.grid(True, alpha = 0.5)
plt.tight_layout()
plt.ylim([0,3])
plt.yticks([0,1,2,3])
plt.show()
```
